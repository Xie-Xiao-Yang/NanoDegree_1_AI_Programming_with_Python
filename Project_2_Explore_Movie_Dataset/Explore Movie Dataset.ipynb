**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_by_year = movie_data.groupby('release_year')[['budget_adj', 'revenue_adj']].mean().sort_values(by='release_year')

movie_by_year['profit_adj'] = movie_by_year['revenue_adj'] - movie_by_year['budget_adj']
```

```python
# plt.bar(movie_by_year.index, movie_by_year['profit_adj'])

plt.plot(movie_by_year.index, movie_by_year['profit_adj'])
plt.xlabel('movie release year')
plt.ylabel("movie's profit(adj)")
plt.title("movie's profit(adj) by release year")
```

分析：
    1. 电影净利润在1980年之前跳动幅度比较大，可能是因为在此之前电影行业预算控制能力不是很成熟
    2. 在1980年之后，电影净利润曲线跳动幅度减少，电影行业预算控制能力逐渐趋于成熟
    3. 电影净利润并没有随着时间推动而增加，可能是因为电影成本也随着越来越高

```python
top10_director = movie_data.groupby('director').count().sort_values(by='id', ascending=False)[:10]

plt.figure(figsize = [15, 25])
index_dir = 0
for d in top10_director.index.values:
    index_dir += 1
    top3_movie = movie_data.query('director == "{}"'.format(d)).sort_values(by='revenue_adj', ascending=False)[:3]
    plt.subplot(5, 2, index_dir)
    plt.bar(top3_movie['original_title'].values, top3_movie['revenue_adj'])
    plt.xlabel('movie name')
    plt.ylabel('revenue(adj)')
    plt.title(d+"'s top 3 revenue movies")
    
```

每部电影票房相差不大的导演： Woody Allen、Martin Scorsese、Steven Spielberg、Ridley Scott <br/>

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
movie_data.head()
```

```python
movie_data['release_year'] = movie_data['release_year'].astype(str)
movie_data.release_date = pd.to_datetime(movie_data['release_date'].str[:-2] + movie_data['release_year'], format='%m/%d/%Y',errors='raise')
movie_between = movie_data[movie_data.release_date.dt.strftime('%Y-%m-%d').between('1968-00-00','2015-06-30')]
movie_count_year = movie_between.groupby('release_year').count().sort_values(by='release_year')['id']
```

```python
plt.figure(figsize = [25, 5])
plt.plot(movie_count_year.index, movie_count_year.values)
plt.xlabel('release year')
plt.ylabel("number of movie release")
plt.title("number of movie release between 1968-00-00 and 2015-06-30")
```

简要分析：<br/>
可以看到随着时间的推移，每年发布的电影基本上都在增长，其中2010~2015年增长速度比以前更快。图中2015年的电影数量减少是因为只统计到2015年6月。

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
def movie_type(mtype):
    is_comedy = ('Comedy'.lower() in mtype.lower())
    is_drama = ('Drama'.lower() in mtype.lower())
    
    if is_comedy and is_drama:
        return 'both'
    elif is_comedy:
        return 'comedy'
    elif is_drama:
        return 'drama'
    else:
        return 'none'
```

```python
movie_between_type = movie_between.set_index('genres').groupby(by=['release_year', movie_type]).count().sort_values(by='release_year')['id']
movie_between_type = movie_between_type.reset_index(1)
movie_between_type.rename(columns={'id':'count'}, inplace=True)
df_both = movie_between_type[movie_between_type['level_1']=='both'].copy().drop(columns='level_1')
df_comedy = movie_between_type[movie_between_type['level_1']=='comedy'].copy().drop(columns='level_1')
df_drama = movie_between_type[movie_between_type['level_1']=='drama'].copy().drop(columns='level_1')
df_comedy = df_comedy + df_both
df_drama = df_drama + df_both
```

```python
plt.figure(figsize = [25, 5])
plt.plot(df_comedy.index, df_comedy['count'], label='comedy')
plt.plot(df_drama.index, df_drama['count'], label='drama')
plt.legend()
plt.xlabel('release year')
plt.ylabel("number of movie release")
plt.title("number of movie release between 1968-00-00 and 2015-06-30")
```

简要分析：<br/>
可以看到随着时间的推移，两种类型的电影数量的增长规律与任务3.4中的大致一样。而在大部分的年份里，drama的电影数量都要高于comedy
